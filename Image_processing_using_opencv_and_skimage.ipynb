```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image=cv2.imread("/content/road.jpeg")
print(image.shape)
print(image)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
#convert the image to grayscale
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
#now we are resizing the image
height=182
width=278
new_height=int(height*0.5)
new_width=int(width*0.5)
resize_img=cv2.resize(gray_image,(new_height,new_width),interpolation=cv2.INTER_AREA)
cv2_imshow(resize_img)
```

```python
#define cordinates of region of interest(ROI)
x=50 #starting x cordinate of the ROI
y=30#starting y cordinate of the ROI
width=170#width of ROI
height=170#height of ROI
#crop the image using ROI coordinates
crop_img=image[y:y+height,x:x+width]
cv2_imshow(crop_img)
```

```python
#Thresholding the image
```

**Thresholding the image**

thresholding is a technique used to seprate pixels into two categories based on a specified threshold value

```python
#thresholding is a technique that generates a binary image,whose pixels have only two values 0 and 1
(thresh,blackandwhiteImage)=cv2.threshold(gray_image,50,255,cv2.THRESH_BINARY)
'''Thresholding is performed with a threshold value of 20.
Pixels with intensity values less than 20 are set to 0 (black), and pixels with
intensity values greater than or equal to 20 are set to 255 (white). '''
cv2_imshow(blackandwhiteImage)
print("...............................................................")
(thresh,blackandwhiteImage)=cv2.threshold(gray_image,80,255,cv2.THRESH_BINARY)
cv2_imshow(blackandwhiteImage)
print("...............................................................")

(thresh,blackandwhiteImage)=cv2.threshold(gray_image,160,255,cv2.THRESH_BINARY)
cv2_imshow(blackandwhiteImage)

print("...............................................................")
(thresh,blackandwhiteImage)=cv2.threshold(gray_image,200,255,cv2.THRESH_BINARY)
cv2_imshow(blackandwhiteImage)



```

#Blurring the image

```python
blur_img=cv2.blur(gray_image,(10,10))
cv2_imshow(blur_img)
```

**GAUSSIAN BLUR**

The blur function uses a simple averaging filter, while the GaussianBlur function applies a Gaussian filter that weights the pixels differently. The GaussianBlur function generally produces better results with smoother blurring and better preservation of image details.

```python
gaus_blur=cv2.GaussianBlur(gray_image,(9,9),10)  #(9,9) represents the kernel size and 10 represents the standard deviation
cv2_imshow(gaus_blur)
```

Rotating and Scaling of the images

```python
#give the rotation angle and scale factor
angle=180
scale=1.0
#get the image dimensions
(h,w)=image.shape[:2]
#calculate center point of the image
center=(w//2,h//2)
#define the rotation matrix
mat=cv2.getRotationMatrix2D(center,angle,scale)
#apply the rotation to imaage
rotate_img=cv2.warpAffine(gray_image,mat,(w,h))
cv2_imshow(rotate_img)
```

**Hough lines transformation**
Hough Line Transformation is a feature extraction technique used in image processing and computer vision to detect straight lines in an image. It is particularly useful in detecting lines in binary images (images with only two colors, typically black and white).and the image should be detected with canny edge

```python
lines=cv2.HoughLinesP(canny1,1,np.pi/180,30)
for line in lines:
  x1,y1,x2,y2=line[0]
  cv2.line(image,(x1,y1),(x2,y2),(0,255,0),4)
cv2_imshow(image)
```

printing the grayscale of the image

```python
from skimage.color import rgb2gray
def compare(image1,image2,title1,title2):
    plt.figure(figsize=(8,3))
    plt.subplot(1,2,1)
    plt.imshow(image1)
    plt.title(title1)
    plt.axis("off")
    plt.subplot(1,2,2)
    plt.imshow(image2,cmap='gray')
    plt.title(title2)
    plt.axis('off')
    plt.show()
image=imread("/content/img1.jpg")
gray=rgb2gray(image)
compare(image,gray,"original image","grayscale image")
```

Flipping the image

```python
img=imread("/content/img1.jpg")
horizontal_flip=np.fliplr(img)
vertical_flip=np.flipud(img)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("original image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(horizontal_flip)
plt.title("horizontal flip image")
plt.axis("off")
plt.tight_layout()
plt.show()
```

Histograms of different colors

```python
def plot_with_hist_channel(image,channel):
  channels=["red","green","blue"]
  channel_idx=channels.index(channel)
  color=channels[channel_idx]
  extracted_channel=image[:,:,channel_idx]
  fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(9,5))
  ax1.imshow(image)
  ax1.axis("off")
  ax2.hist(extracted_channel.ravel(),bins=256,color=color)
  ax2.set_xlim([0,100])
  ax2.set_title(f'{channels[channel_idx]} histogram')
```

```python
color_img=imread("/content/color.jpg")
plot_with_hist_channel(color_img,"blue")
```

```python
color_img=imread("/content/color1.jpg")
plot_with_hist_channel(color_img,'green')
```
